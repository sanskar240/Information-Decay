```python
import pandas as pd
import torch
from sentence_transformers import SentenceTransformer

# Load the MATRES dataset
matres_file = 'MATRES.csv'  # Replace with your file path
matres_df = pd.read_csv(matres_file)

# Display the columns in the dataset
print("Columns in the dataset:", matres_df.columns.tolist())

# Prepare sentences from the 'bodytext' column (assuming it exists)
sentences = matres_df['bodytext'].dropna().tolist()

# Include an example sentence
example_sentence = "This is an example sentence."
sentences.append(example_sentence)

# Load SBERT model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Generate embeddings
with torch.no_grad():  # Disable gradient tracking for inference
    embeddings = model.encode(sentences)

# Print the shape of the embeddings
print("Shape of embeddings:", embeddings.shape)

# Print the embedding for the example sentence
example_embedding = embeddings[-1]  # Get the last embedding for the example sentence
print("Embedding for the example sentence:", example_embedding)

# Save embeddings to a file (optional)
torch.save(embeddings, 'sentence_embeddings.pt')
```
