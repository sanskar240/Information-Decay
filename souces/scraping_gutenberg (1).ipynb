```python
import nltk
nltk.download('punkt')
```

```python
from bs4 import BeautifulSoup
import spacy
from nltk.tokenize import sent_tokenize
import re

# Load spaCy model for Named Entity Recognition
nlp = spacy.load('en_core_web_sm')

# Function to extract and process text from a single HTML file
def extract_sentences_from_file(file_path):
    # Read the HTML file
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Parse HTML content
    soup = BeautifulSoup(content, 'lxml')

    # Get all text
    all_text = soup.get_text()

    # Clean text: remove copyright notes, annotations, and unwanted characters
    cleaned_text = clean_text(all_text)

    # Tokenize into sentences
    sentences = sent_tokenize(cleaned_text)

    # Filter sentences based on temporal expressions and length
    filtered_sentences = [
        sentence for sentence in sentences if detect_events_and_temporals(sentence) and len(sentence.split()) <= 20
    ]

    return filtered_sentences

# Function to clean the extracted text
def clean_text(text):
    # Remove copyright notices
    text = re.sub(r'\[Copyright.*?\]', '', text)
    # Remove anything in curly braces
    text = re.sub(r'\{.*?\}', '', text)
    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

# Function to detect temporal expressions and events using spaCy
def detect_events_and_temporals(sentence):
    doc = nlp(sentence)
    for ent in doc.ents:
        if ent.label_ in ['DATE', 'EVENT']:
            return True
    return False

# Main logic to process the specified HTML files
all_filtered_sentences = []

# List of your HTML files (now includes data3.html)
file_paths = ['data.html', 'data2.html', 'data3.html']

for file_path in file_paths:
    filtered_sentences = extract_sentences_from_file(file_path)
    all_filtered_sentences.extend(filtered_sentences)

# Store all filtered sentences in a single output file
with open('filtered_sentences.txt', 'w', encoding='utf-8') as f:
    for sentence in all_filtered_sentences:
        f.write(f"{sentence}\n")

# Output the number of sentences extracted
number_of_sentences = len(all_filtered_sentences)
print(f"Total number of sentences extracted from all files: {number_of_sentences}")

# Optional: Output the filtered sentences to console
for sentence in all_filtered_sentences:
    print(sentence)

```
